```python
# IMPORTING ALL MODULES

import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus
```

```python
# IMPORTING THE DATA

iris=datasets.load_iris()
df_x=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df_x.head(5))
print()
df_y=iris.target
print(df_y)
```

```python
# TRAIN TEST SPLIT

x_train,x_test,y_train,y_test=train_test_split(df_x,df_y)
```

```python
# DECISION TREE CRITERION GINI INDEX

decision_tree_gini=DecisionTreeClassifier(criterion='gini',random_state=0)
decision_tree_gini=decision_tree_gini.fit(x_train,y_train)
y_train_gini=decision_tree_gini.predict(x_train)
y_test_gini=decision_tree_gini.predict(x_test)

print("Accuracy test :",metrics.accuracy_score(y_train,y_train_gini))
print("Accuracy test :",metrics.accuracy_score(y_test,y_test_gini))
print()

dot_data=StringIO()
export_graphviz(decision_tree_gini,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=df_x.columns,class_names=['0','1','2'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('gini_index.png')
Image(graph.create_png())
```

```python
# DECISION TREE CRITERION LOG LOSS

decision_tree_log_loss=DecisionTreeClassifier(criterion='log_loss',random_state=0)
decision_tree_log_loss=decision_tree_log_loss.fit(x_train,y_train)
y_train_log_loss=decision_tree_log_loss.predict(x_train)
y_test_log_loss=decision_tree_log_loss.predict(x_test)

print("Accuracy test :",metrics.accuracy_score(y_train,y_train_log_loss))
print("Accuracy test :",metrics.accuracy_score(y_test,y_test_log_loss))
print()

dot_data=StringIO()
export_graphviz(decision_tree_log_loss,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=df_x.columns,class_names=['0','1','2'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('log_loss.png')
Image(graph.create_png())
```

```python
# DECISION TREE CRITERION ENTROPY

decision_tree_entropy=DecisionTreeClassifier(criterion='entropy',random_state=0)
decision_tree_entropy=decision_tree_entropy.fit(x_train,y_train)
y_train_entropy=decision_tree_entropy.predict(x_train)
y_test_entropy=decision_tree_entropy.predict(x_test)

print("Accuracy test :",metrics.accuracy_score(y_train,y_train_entropy))
print("Accuracy test :",metrics.accuracy_score(y_test,y_test_entropy))
print()

dot_data=StringIO()
export_graphviz(decision_tree_entropy,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=df_x.columns,class_names=['0','1','2'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('entropy.png')
Image(graph.create_png())
```
